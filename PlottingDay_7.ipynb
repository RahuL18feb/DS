```python
plt.figure(figsize=(6 , 4))
plt.bar(gender_counts.index , gender_counts.values ,
        color = ["blue" , "red"] , alpha = 0.7 , edgecolor = "black")
plt.title("Gender counts")
plt.xlabel("Gender")
plt.ylabel("No. of Passenger")
```

```python
plt.figure(figsize=(6 , 4))
plt.bar(gender_counts.index , gender_counts.values ,
        color = ["blue" , "red"] , alpha = 0.7 , edgecolor = "black")
plt.title("Gender counts")
plt.xlabel("Gender")
plt.ylabel("No. of Passenger")
plt.tight_layout()
```

```python
plt.figure(figsize=(6 , 4))
plt.bar(gender_counts.index , gender_counts.values ,
        color = ["blue" , "red"] , alpha = 0.7 , edgecolor = "black")
plt.title("Gender counts")
plt.xlabel("Gender")
plt.ylabel("No. of Passenger")
plt.tight_layout()

plt.savefig("Gender_counts.jpeg")
```

#Horizontal bar

#Histogram

```python
titanic["age"].isnull().sum()
```

```python
ages = titanic["age"].dropna()
```

```python
ages.isnull().sum()
```

```python
plt.figure(figsize = (8,5))
plt.hist(ages)
```

```python
plt.figure(figsize = (8,5))
plt.hist(ages , edgecolor = "black")
```

```python
plt.figure(figsize = (8,5))
plt.hist(ages , bins = 30 , edgecolor = "black")
```

```python
plt.figure(figsize = (8,5))
plt.hist(ages , bins = 30 , edgecolor = "black" , color = "skyblue")
```

```python
plt.figure(figsize = (8,5))
plt.hist(ages , bins = 30 , edgecolor = "black")
plt.title("Age Distribution")
plt.xlabel("age")
plt.ylabel("frequency")
plt.tight_layout()
plt.grid(True)
```

#Pie Chart

```python
survived_counts = titanic["survived"].value_counts()
```

```python
survived_counts  # 0 - death , 1 - survived
```

```python
plt.figure(figsize=(6 , 6))
plt.pie(survived_counts)
```

```python
plt.figure(figsize=(6 , 6))
plt.pie(survived_counts , colors = ["red" , "green"])
```

```python
plt.figure(figsize=(6 , 6))
plt.pie(survived_counts , labels = ["Died" , "survived"] , colors = ["red" , "green"])
```

```python
plt.figure(figsize=(6 , 6))
plt.pie(survived_counts , labels = ["Died" , "survived"] ,
        colors = ["red" , "green"] , autopct="%1.1f%%")
```

```python
plt.figure(figsize=(6 , 6))
plt.pie(survived_counts , labels = ["Died" , "survived"] ,
        colors = ["red" , "green"] , autopct="%1.2f%%")
```

```python
plt.figure(figsize=(6 , 6))
plt.pie(survived_counts , labels = ["Died" , "survived"] ,
        colors = ["red" , "green"] , autopct="%1.1f%%" ,
        startangle=90)
```

```python
plt.figure(figsize=(6 , 6))
plt.pie(survived_counts , labels = ["Died" , "survived"] ,
        colors = ["red" , "green"] , autopct="%1.1f%%")
plt.title("Survival Percentage")
plt.legend()
```

#Scatter Plot - Bivariate- Num vs Num

```python
titanic[["age" , "fare"]].isnull().sum()
```

```python
data = titanic.dropna(subset = ["age" , "fare"])
```

```python
data[["age" , "fare"]].isnull().sum()
```

```python
data[["age" , "fare"]].shape
```

```python
data[["age" , "fare"]].notnull().sum()
```

```python
data.head()
```

```python
plt.figure(figsize = (8,6))
plt.scatter(data["age"] , data["fare"]) #Scatter
```

```python
plt.figure(figsize = (8,6))
plt.scatter(data["age"] , data["fare"],
            color = "skyblue" , alpha = 0.5 , marker = "s")
```

```python
plt.figure(figsize = (8,6))
plt.scatter(data["age"] , data["fare"],
            color = "orange" , alpha = 0.5)
plt.title("age - fare")
plt.xlabel("age")
plt.ylabel("fare")
```
