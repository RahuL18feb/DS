#Seaborn library
-SEABORN library is based on matplotlib

- it is built on the top of matplotlib

-very useful and simpler for visualization

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
titanic = sns.load_dataset("titanic")
```

```python
titanic.head(1)
```

#count plot

```python
titanic['class'].value_counts()
```

```python
plt.figure(figsize = (6,6))
sns.countplot(data = titanic, x = "class")
```

```python
plt.figure(figsize = (6,6))
sns.countplot(data = titanic, y = "class" , color = "red")
```

```python
plt.figure(figsize = (6,6))
sns.countplot(data = titanic, x = "class" , palette = "Set1")
```

```python
plt.figure(figsize = (6,6))
sns.countplot(data = titanic, x = "class" , hue = "sex" , palette = "Set1")
```

```python
plt.figure(figsize = (6,6))
sns.countplot(data = titanic, x = "class" , hue = "sex" , palette = "Set1" , edgecolor = "black")
```

#Bar plot

```python
plt.figure(figsize = (6,4))
sns.barplot(data = titanic , x = "class" , y = "age")
```

```python
plt.figure(figsize = (6,4))
sns.barplot(data = titanic , x = "class" , y = "age" , palette = "Set2")
```

```python
plt.figure(figsize = (6,4))
sns.barplot(data = titanic , x = "class" , y = "age" , hue = "sex" , palette = "Set2")
```

BOX PLOT - outlier detection

```python
sns.boxplot(data = titanic , x ="class" , y = "fare")
```

```python
sns.boxplot(data = titanic , x ="class" , y = "fare" , hue = "sex" , palette = "viridis")
```

#VIOLIN plot

```python
sns.violinplot(data = titanic , x ="class" , y = "fare" , palette = "viridis")
```

#HISTOGRAM - continuous feature distribution

```python
sns.histplot(data = titanic , x = "age", bins = 30)
```

```python
sns.histplot(data = titanic , x = "age", bins = 30 , hue = "sex")
```

```python
sns.histplot(data = titanic , x = "age", bins = 30 , hue = "sex" , alpha = 0.6)
```

```python
sns.histplot(data = titanic , x = "age", bins = 30 , palette = "red", kde = "True"  , alpha = 0.6)
```

KDE plot - kernal density estimator

```python
sns.kdeplot(data = titanic, x = "age")
```

```python
sns.kdeplot(data = titanic, x = "age" , shade = True)
```

```python
sns.kdeplot(data = titanic, x = "age" , fill = True)
```

#Heatmap - correlation matrix

```python
correlation_matrix = titanic.corr(numeric_only = True)
```

```python
correlation_matrix
```

```python
sns.heatmap(correlation_matrix)
```

```python
sns.heatmap(correlation_matrix , annot = True)
```

```python
sns.heatmap(correlation_matrix , annot = True , cmap ="coolwarm")
```

#Scatter plot - Numerical vs Numerical

```python
sns.scatterplot(data = titanic , x = "age" , y ="fare" , marker = "^")
```

#regression plot

```python
sns.regplot(data = titanic , x = "age", y='fare' , line_kws = dict(color='r'))
```

#subplot

```python
plt.figure(figsize = (14,6))

plt.subplot(1,2,1)
sns.scatterplot(data = titanic , x = "age" , y = "fare")


plt.subplot(1,2,2)
sns.kdeplot(data = titanic , x = "fare")
```

```python
plt.figure(figsize = (14,6))

plt.subplot(1,2,2)
sns.scatterplot(data = titanic , x = "age" , y = "fare")


plt.subplot(1,2,1)
sns.kdeplot(data = titanic , x = "fare")
```

```python
plt.figure(figsize = (14,6))

plt.subplot(1,2,2)
sns.scatterplot(data = titanic , x = "age" , y = "fare")


plt.subplot(2,2,2)
sns.scatterplot(data = titanic , x = "age" , y = "fare")


plt.subplot(2,2,1)
sns.kdeplot(data = titanic , x = "fare")
```

#Pair plot

```python
sns.pairplot(titanic[["age", "fare" , "pclass" , "survived"]])
```
