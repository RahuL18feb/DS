```python
#age_bins = [0,12,18,35,60,titanic["age"].max()]
```

```python
age_bins = [0,12,18,35,60,80]
age_labels=["child", "teens", "young", "adult", "senior"]

titanic["age_group"] = pd.cut(titanic["age"], bins = age_bins, labels = age_labels)
```

```python
titanic[["age", "age_group"]].head()
```
